<h1>  Missing Value Treatment </h1>

<h2> Fill missing 'Age' with median </h2>

```python
titanic['Age'].fillna(titanic['Age'].median(), inplace=True)
```

<h2> Fill missing 'Embarked' with mode </h2>

<h1> Outlier Detection </h1>

```python
# Using Boxplots to Detect Outliers
for col in ['Age', 'Fare']:
    plt.figure(figsize=(8,4))
    sns.boxplot(titanic[col])
    plt.title(f'Boxplot of {col}')
    plt.show()


print("🔎 Observation: Age has a few outliers among very young and very old passengers. Fare has significant high-end outliers with some passengers paying very high ticket prices.")
```

<h2> Capping Fare Outliers </h2>

```python
q1 = titanic['Fare'].quantile(0.25)
q3 = titanic['Fare'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
titanic['Fare'] = np.where(titanic['Fare'] > upper_bound, upper_bound, titanic['Fare'])
```

<h1> Feature Engineering </h1>

```python
# Create New Feature: FamilySize
titanic['FamilySize'] = titanic['SibSp'] + titanic['Parch'] + 1
```

```python
# Create IsAlone Feature
titanic['IsAlone'] = 1  # Initialize to 1 (means alone)
titanic['IsAlone'].loc[titanic['FamilySize'] > 1] = 0
```

```python
# Create Title Feature from Name
titanic['Title'] = titanic['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
```

```python
# Group rare titles
rare_titles = titanic['Title'].value_counts()[titanic['Title'].value_counts() < 10].index
titanic['Title'] = titanic['Title'].replace(rare_titles, 'Rare')
```

<h1>  ANOVA Test (Categorical vs Target) </h1>


```python
import scipy.stats as stats
```

<h2> Gender vs Survived </h2>

```python
table = pd.crosstab(titanic['Sex'], titanic['Survived'])
chi2, p, dof, expected = stats.chi2_contingency(table)
print(f'Gender p-value: {p:.4f}')
```

<h2> Pclass vs Survived </h2>

```python
table = pd.crosstab(titanic['Pclass'], titanic['Survived'])
chi2, p, dof, expected = stats.chi2_contingency(table)
print(f'Pclass p-value: {p:.4f}')
```

<h2> Embarked vs Survived </h2>

```python
table = pd.crosstab(titanic['Embarked'], titanic['Survived'])
chi2, p, dof, expected = stats.chi2_contingency(table)
print(f'Embarked p-value: {p:.4f}')
```

<h2> </h2>

<h2> Check missing values </h2>

```python
td.isnull().sum()
```

```python
td.isnull().sum().sort_values(ascending = False)
```

<h2> Fill missing Age with median </h2>

```python
td['Age'].fillna(td['Age'].median(), inplace=True)
```

<h2> Fill missing Fare with median </h2>

```python
td['Fare'].fillna(td['Fare'].median(), inplace=True)
```

<h2> Drop Cabin </h2>

```python
td.drop('Cabin', axis=1, inplace=True)
```

<h1> Feature Engineering on Test Data </h1>

```python
# Family Size
td['FamilySize'] = td['SibSp'] + td['Parch'] + 1
```

```python
# Is Alone
td['IsAlone'] = 1
td['IsAlone'].loc[td['FamilySize'] > 1] = 0
```

```python
# Title Extraction
td['Title'] = td['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
rare_titles = ['Dr', 'Rev', 'Col', 'Major', 'Mlle', 'Countess', 'Ms', 'Lady', 'Dona', 'Jonkheer', 'Sir', 'Don', 'Capt']
td['Title'] = td['Title'].replace(rare_titles, 'Rare')
```

<h1> Encode Categorical Variables </h1>

```python
# Encode 'Sex' column manually
titanic['Sex'] = titanic['Sex'].apply(lambda x: 0 if x == 'male' else 1)
td['Sex'] = td['Sex'].apply(lambda x: 0 if x == 'male' else 1)
```

```python
# Encode 'Embarked' column manually
titanic['Embarked'] = titanic['Embarked'].apply(lambda x: 0 if x == 'S' else (1 if x == 'C' else 2))
td['Embarked'] = td['Embarked'].apply(lambda x: 0 if x == 'S' else (1 if x == 'C' else 2))
```

```python
# Encode 'Title' column manually
titanic['Title'] = titanic['Title'].apply(lambda x: 0 if x == 'Mr' else (1 if x == 'Miss' else (2 if x == 'Mrs' else (3 if x == 'Master' else 4))))
td['Title'] = td['Title'].apply(lambda x: 0 if x == 'Mr' else (1 if x == 'Miss' else (2 if x == 'Mrs' else (3 if x == 'Master' else 4))))
```

<h1> Select Features for Modeling </h1>

```python
features = ['Pclass', 'Sex', 'Age', 'Fare', 'Embarked', 'FamilySize', 'IsAlone', 'Title']
X_train = titanic[features]
y_train = titanic['Survived']

X_test = td[features]
```

<h1> Build and Train Logistic Regression Model </h1>

```python
from sklearn.linear_model import LogisticRegression

# Create model
model = LogisticRegression()

# Train model
model.fit(X_train, y_train)
```

<h1> Predict on Test Set </h1>

```python
# Predict survival for test set
y_pred = model.predict(X_test)
```

<h1> Submission File </h1>

```python
submission = pd.DataFrame({
    'PassengerId': td['PassengerId'],
    'Survived': y_pred
})

# Save to CSV
submission.to_csv('submission.csv', index=False)
```
